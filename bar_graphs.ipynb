```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('recent-grads.csv')
```

```python
data.head()
```

```python
data.iloc[:, :-1]
data.iloc[:, :-2]
```

```python
#Top 5 majors 
top_five_majors = data.sort_values(by="median", ascending=False).head()
```

```python
top_five_majors.head()
```

```python
top_five_majors.plot.barh(x="major",y="median",color="red",rot=0,fontsize=8)
plt.title("Median Salary from top 5 highest paying majors")
plt.ylabel("Median Salary ($USD)")
plt.xlabel("Major")
#plt.xticks(rotation=45)
#plt.yticks(rotation=90)
plt.show()
```

```python
low_5_majors= data.sort_values(by="median").head()
```

```python
low_5_majors.head()
```

```python
low_5_majors.plot.barh(x="major",y=["total","women","men"],stacked=False,rot=0)
plt.title("Gender distribution across the lowest paying majors")

plt.show()
```

```python
low_5_majors.plot.barh(x="major",y=["men","women"],color=["blue","orange"], stacked=True)
plt.title("Gender Distributions ")
```

```python
data.plot.scatter(x="non_college_jobs",y="p75th")
plt.show()
```

```python
cat_sums = low_5_majors.groupby("major_category")["total"].sum().sort_values()
cat_sums.head()
```

```python
cat_sums.plot.pie(autopct='%1.1f%%', shadow=True)
plt.show()
```

```python
data.boxplot(column="median",by="major_category",grid=False,rot=90)
plt.show()
```
